```python
import json
import stomp
import atexit
import time

class MessageListener(stomp.ConnectionListener):
    def __init__(self, topic_name):
        self.topic_name = topic_name

    def on_message(self, sent_info,message):
        try:
            json_message = json.loads(message)
            # Process the JSON message
            print(f"Received JSON message '{json_message}' from topic '{self.topic_name}'")
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON message: {e}")

class Subscriber:
    def __init__(self, topic_nameP, idP , headersP):
        self.conn = stomp.Connection([('amq', 61613)])  
        self.conn.set_listener('', MessageListener(topic_nameP))
        self.conn.connect()
        if headersP != "":
            self.conn.subscribe(destination=topic_nameP, id=idP, ack='auto', headers=headersP)
        else:
            self.conn.subscribe(destination=topic_nameP, id=idP, ack='auto')
         
    def disconnect(self):
        self.conn.disconnect()

    def connect(self):
        self.conn.connect()
```

```python
header1 = {'activemq.subscriptionName': 'sub1', 'durable': 'true','selector':"data = 'weather'" }
subscriber1 = Subscriber('/topic/Atlanta', 'sub1', header1)
```

```python
subscriber1.disconnect()
```
